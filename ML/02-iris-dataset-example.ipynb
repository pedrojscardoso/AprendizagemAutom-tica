# Iris dataset

The Iris dataset is a classic and very easy multi-class classification dataset. It consists of 3 different types of irises’ (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray. The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Petal Width.

The iris dataset is a classic and very easy multi-class classification dataset.

```python
# Load libraries
from sklearn import datasets
import pandas as pd
```

Load the data from the `sklearn` dataset

```python
iris = datasets.load_iris()
```

Print information about the dataset

```python
print(iris.DESCR)
```

The data is an array of 150 instances

```python
iris.data
```

Each instance as 4 features (given in the following array)

```python
iris.feature_names
```

The target array values corresponds a value to each instance in the `data`

```python
iris.target
```

The names of the target's are in the `target_name` property. E.g., iris 'setosa' correponds to the target 0.

```python
iris.target_names
```

Of course we can access a single instance by accessing the array

```python
iris.data[0]
```

## use of Pandas
You can also make a pandas dataframe with iris data and perform your analysis from there

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
df
```

```python
df.plot(figsize=(20,10))
```

```python
df.groupby(by='target').plot()
```
